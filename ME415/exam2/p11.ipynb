```python
%reset -f
%load_ext autoreload
%autoreload 2

import numpy as np
from matplotlib import pyplot as plt

Kv = 62.831853 #(rad/s)/volt
R = 1 #ohm
I0 = 0.2 #amp
D = 0.25 #m
V = 12 #m/s
density = 1 #kg/m^3
Vs = 10 #volts

rotation_speed = np.linspace(0,1000,1000) #rad/s
n = rotation_speed / (2 * np.pi)
J = V / (n * D)

CT = 0.1 - 0.04 * J - 0.07 * J**2
CQ = 0.005 + 0.01 * J - 0.02 * J**2
Qprop = CQ * density * n**2 * D**5

I = (Vs - rotation_speed / Kv) * (1 / R)
Qmotor = (1/Kv) * (I - I0)

plt.plot(rotation_speed,Qprop,c='red')
plt.plot(rotation_speed,Qmotor,c='blue')
plt.axvline(515,c='black')
plt.show()
```
